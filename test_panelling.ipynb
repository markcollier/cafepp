```python
print('BEGIN')

# https://stackoverflow.com/questions/13784201/matplotlib-2-subplots-1-colorbar

import numpy as np

def plot_try(x,y,z,fig,ax,nrowcols):
# add_axes: rect [left, bottom, width, height] 
  import matplotlib.pyplot as plt

  cs = ax.contourf(x,y,z,[0,.25,.5,1,2,3,5])

  if(nrowcols==2): #single panel
    cb = plt.colorbar(cs)
  else: #multiple panel, attach colorbar to figure.
    fig.subplots_adjust(right=0.9)
    cbar_ax = fig.add_axes([0.95, 0.10, 0.04, 0.8])
    fig.colorbar(cs, cax=cbar_ax)
  
  return()
  
x, y = np.meshgrid(np.arange(7), np.arange(10))
z1 = np.sin(0.2 * x) * np.cos(0.52 * y) *1
z2 = np.sin(0.4 * x) * np.cos(0.52 * y) *10
z3 = np.sin(0.6 * x) * np.cos(0.52 * y) *100
z4 = np.sin(0.8 * x) * np.cos(0.52 * y) *1000

# print(z1.shape)
# print(np.min(z1),np.min(z2),np.min(z3),np.min(z4))
# print(np.max(z1),np.max(z2),np.max(z3),np.max(z4))

%matplotlib inline


nrows=2
ncols=2
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(figsize=(10, 5), nrows=nrows, ncols=ncols)
nrowcols=nrows+ncols
plot_try(x,y,z1,fig,ax1,nrowcols)
plot_try(x,y,z2,fig,ax2,nrowcols)
plot_try(x,y,z3,fig,ax3,nrowcols)
plot_try(x,y,z4,fig,ax4,nrowcols)

nrows=1
ncols=2
fig, (ax1,ax2) = plt.subplots(figsize=(5, 5), nrows=nrows, ncols=ncols)
nrowcols=nrows+ncols
plot_try(x,y,z1,fig,ax1,nrowcols)
plot_try(x,y,z2,fig,ax2,nrowcols)

nrows=1
ncols=2
fig, (ax1,ax2) = plt.subplots(figsize=(5, 2.5), nrows=nrows, ncols=ncols)
nrowcols=nrows+ncols
plot_try(x,y,z1,fig,ax1,nrowcols)
plot_try(x,y,z2,fig,ax2,nrowcols)

nrows=1
ncols=1
fig, ax = plt.subplots(nrows=nrows, ncols=ncols)
nrowcols=nrows+ncols
plot_try(x,y,z2,fig,ax,nrowcols)

print('END')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data to plot.
x, y = np.meshgrid(np.arange(7), np.arange(10))
z = np.sin(0.5 * x) * np.cos(0.52 * y)


%matplotlib inline


fig, axes = plt.subplots(nrows=2, ncols=2)
for ax in axes.flat:
#     im = ax.imshow(np.random.random((10,10)), vmin=0, vmax=1)
    im = ax.contourf(x,y,z)

fig.subplots_adjust(right=0.8)
cbar_ax = fig.add_axes([0.85, 0.15, 0.05, 0.7])
fig.colorbar(im, cax=cbar_ax)

plt.show()
```
